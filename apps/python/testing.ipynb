```python
import os
import importlib.util

current_directory = os.getcwd()
module_path = os.path.abspath(os.path.join(current_directory, '..', '..', 'install', 'bin', 'my_python_package.cpython-313-darwin.so'))
module_name = 'my_python_package'
 

module_spec = importlib.util.spec_from_file_location(module_name, module_path)
 
my_python_package = importlib.util.module_from_spec(module_spec)
module_spec.loader.exec_module(my_python_package)
```

```python
a = my_python_package.Structure(1,2)
b = my_python_package.Operation()
z = b(a)
```
